```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

1. Read the input file into a dataframe
2. The multi line message have none values, remove them


```python
df = pd.read_csv('WhatsApp Chat with Anaga.txt',sep='M -', header=None,names=['Time','Message'])
s1 = df.Message.isnull().sum()
df = df.dropna(subset=['Message'])
df[['Sender','MessageText']] = df.Message.str.split(":",expand=True,n=1) 
df.drop(['Message'],axis=1,inplace=True)

```

```python
df.head()
df.info()
df['Time']= pd.to_datetime(df['Time'])
df.shape
df.dropna(inplace=True)
df.head()
```

Number of messages by each person

```python
Sendercount=df[['MessageText','Sender']].groupby(['Sender']).size()

sns.set()
plt.figure(figsize=(20,5))

Sendercount.sort_values(ascending=True,inplace=True)
sendcountchart=sns.barplot(Sendercount.index,Sendercount.values)
sendcountchart.tick_params(labelsize=14)
sendcountchart.set_xlabel(xlabel="Sender",fontsize=20)
sendcountchart.set_title('PING COUNT',fontsize=25)
```

Lets plot the message count on day

```python

df.index = df['Time'] 

names=df.Sender.unique()
indicount=[]

sns.set_style("dark")
dfindi=df[df.Sender.str.contains('vis',case=False)]
daycount=dfindi.resample('D').count()
hourcount=dfindi.groupby(dfindi.index.hour).count()

a_row = pd.Series([0,0,0])
for x in range(24):
    if x not in hourcount.index:
        pass
        zero_hours = pd.DataFrame([a_row], index = [x])
        hourcount = pd.concat([zero_hours, hourcount])
        

#add missing hours as zero
#print(hourcount)
hourcount.drop(['Time', 'Sender'], axis=1,inplace=True)
hourcount=hourcount.sort_index(axis = 0) 
sns.set()
plt.figure(figsize=(20,5))
hourcountchart=sns.lineplot(x=hourcount.index,y=hourcount['MessageText'], linewidth=5)
#hourcountchart.set_xticklabels(hourcount.index)
hourcountchart.set(xticks=hourcount.index)
hourcountchart.tick_params(labelsize=14)
hourcountchart.set_xlabel(xlabel="Hour",fontsize=20)
hourcountchart.set_ylabel(ylabel="Number of Messages",fontsize=20)
hourcountchart.set_title('ab',fontsize=25)
plt.fill_between(hourcount.index, hourcount['MessageText'])


#Daycount
print(daycount)
daycount.drop(['Time', 'Sender'], axis=1,inplace=True)
daycount=daycount.sort_index(axis = 0) 
sns.set()
plt.figure(figsize=(20,5))
daycountchart=sns.scatterplot(x=daycount.index,y=daycount['MessageText'])
daycountchart.xaxis_date()
#hourcountchart.set_xticklabels(hourcount.index)
daycountchart.set(xticks=hourcount.index)
daycountchart.tick_params(labelsize=14)
daycountchart.set_xlabel(xlabel="DAY",fontsize=20)
daycountchart.set_ylabel(ylabel="Number of Messages",fontsize=20)
daycountchart.set_title('ab',fontsize=25)
#plt.fill_between(daycountchart.index, hourcount['MessageText'])
```
